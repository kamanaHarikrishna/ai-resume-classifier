```python


# Step 3: Random Oversampling
# Find the size of the majority class
max_size = df['Category'].value_counts().max()

# Oversample each category to match the majority class
dfs = []
for category in df['Category'].unique():
    df_cat = df[df['Category'] == category]
    df_cat_upsampled = resample(df_cat, 
                                 replace=True,     # sample with replacement
                                 n_samples=max_size,    # to match majority class
                                 random_state=42)
    dfs.append(df_cat_upsampled)

# Step 4: Combine all upsampled categories into a balanced dataset
df_balanced = pd.concat(dfs)

# Step 5: Shuffle the dataset
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Step 6: Print new balanced distribution
print("\nBalanced Category Distribution:\n", df_balanced['Category'].value_counts())
```

```python
import re
import string

def clean_resume(text):
    # 1. Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # 2. Remove mentions and hashtags
    text = re.sub(r'@\w+|#\w+', '', text)

    # 3. Remove special characters and punctuation
    text = re.sub(r'[{}]'.format(re.escape(string.punctuation)), ' ', text)

    # 4. Remove non-alphanumeric characters (except spaces)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    # 5. Remove extra whitespaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df['Resume_cleaned'] = df['Resume_str'].apply(clean_resume)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib

# Initialize vectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Fit the vectorizer on the cleaned resumes
tfidf.fit(df['Resume_cleaned'])

# Transform resumes into TF-IDF matrix
X = tfidf.transform(df['Resume_cleaned'])

# Print the shape
print(f"TF-IDF matrix shape: {X.shape}")

# ✅ Save the trained vectorizer to a file
joblib.dump(tfidf, 'TfidfVectorizer.pkl')
```

```python
from sklearn.model_selection import train_test_split

# Feature matrix (X) and target variable (y)
X = tfidf.fit_transform(df['Resume_cleaned']).toarray()
y = df['Category']

# Split into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,         # 20% test data
    random_state=42,       # for reproducibility
    stratify=y             # ensures class balance in both sets
)

# Output shape for confirmation
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
import joblib

# Initialize the model
model = MultinomialNB()

# Train the model
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", round(accuracy * 100, 2), "%")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# ✅ Save the trained model
joblib.dump(model, 'resume_classifier.pkl')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, accuracy_score

# Store results
results = {}

# Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
lr_preds = lr.predict(X_test)
results['Logistic Regression'] = {
    'accuracy': accuracy_score(y_test, lr_preds),
    'report': classification_report(y_test, lr_preds, output_dict=True)
}

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)
results['Random Forest'] = {
    'accuracy': accuracy_score(y_test, rf_preds),
    'report': classification_report(y_test, rf_preds, output_dict=True)
}

# Linear SVM
svm = LinearSVC()
svm.fit(X_train, y_train)
svm_preds = svm.predict(X_test)
results['Linear SVM'] = {
    'accuracy': accuracy_score(y_test, svm_preds),
    'report': classification_report(y_test, svm_preds, output_dict=True)
}

# Return summarized results
import pandas as pd
summary = pd.DataFrame({
    model: {
        'Accuracy': round(metrics['accuracy'] * 100, 2)
    } for model, metrics in results.items()
}).T

summary
```
